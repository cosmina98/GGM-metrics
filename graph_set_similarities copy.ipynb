```python
#pos and ne

current_smiles_datasets=['ames', 'bbb_martins', 'cyp1a2_veith', 'cyp2c19_veith','herg_karim','lipophilicity_astrazeneca']
splits=['_25','_33','_40','_50']
#device=torch.device('cpu')
structural_statistic=['nspdk','WL','common_substructures']
metrics_type=['nn', 'structural','molecular']
generators=['wgan','moflow','hiervae','stgg', 'gdss','jtnn']
metrics={}
name_decomposition_list = []
name_decomposition_list.append(['node_edge',atom()])
name_decomposition_list.append(['unlabelled_graph_cycle',cycle(abstraction_level='unlabelled_graph_process')])
name_decomposition_list.append(['cycle', cycle()])
name_decomposition_list.append(['neighborhood r=1', neighborhood(size=1)])
name_decomposition_list.append(['neighborhood r=2', neighborhood(size=2)])
#name_decomposition_list.append(['pair neighborhood r=2 d=0:3', pairwise_neighborhood(size=2, min_distance=0, max_distance=3)])
metrics={}
for generator in generators:
  for dataset_name in current_smiles_datasets:
          for split in splits:
                #if (dataset_name+split =='cyp1a2_veith_33') and ((generator=='hiervae') or (generator=='moflow')) or dataset_name+split =='cyp1a2_veith_25':
                    #print(dataset_name+split,generator)
                    #continue
                  print(dataset_name+split,generator)
                  train1_graphs , train1_targets,train2_graphs , train2_targets, test_graphs, test_targets =get_data(dataset_name+split, return_smiles=False)
                  #pos_train1_graphs=np.array(train1_graphs)[[i for (i,t) in enumerate(train1_targets) if t!=0]]
                  #print(len(train1_graphs),len(set(train1_graphs)))
                  generated_graphs, generated_targets=get_generated_data(dataset_name+split, generator_name=generator)
                  #pos_generated_graphs=np.array(generated_graphs)[[i for (i,t) in enumerate(generated_targets) if t!=0]]
                  #neg_generated_graphs=np.array(generated_graphs)[[i for (i,t) in enumerate(generated_targets) if t==0]]
                  for name, df in name_decomposition_list:
                      start = time.time()
                      score = graph_set_similarity(train1_graphs, generated_graphs, decomposition_function=df)
                      elapsed = time.time() - start
                      print('%30s: %5.3f   [%4.1f s  (%4.1f m)]'%(name, score, elapsed, elapsed/60))
                      metrics[dataset_name+split+" "+generator+ " " + name +'_similarity']= score
                      metrics[dataset_name+split+" "+generator+ " " + name+ "_similarity_time" ]= elapsed
                      print(metrics)
                    
```
